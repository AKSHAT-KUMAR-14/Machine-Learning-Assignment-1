## Simple Linear Regression 

```python
import numpy as np 
import pandas as pd 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
test=pd.read_csv('data/Linear Regression - Sheet1.csv')
```

## EDA

```python
print(test.info())
print(test.head())
print(test.shape)
print(test.describe())
```

### Following model has on one feature so it has 1:1 ration between input and predictions so it is called Linear Regression.

```python
import matplotlib.pyplot as plt
test["Y"].hist()
plt.xlabel="X"
plt.ylabel="Y"
plt.show()
```

## Data Cleaning

```python
test = test.fillna(test.median(numeric_only=True))
```

## Feature Extraction

```python
Y=test["Y"]
X=test.drop(columns=["Y"])

XT,XV,YT,YV=train_test_split(X,Y,test_size=0.2,random_state=42)
```

## Model Training

```python
model=LinearRegression()
model.fit(XT,YT)
```

```python
preds=model.predict(XV)
rmse=np.sqrt(mean_squared_error(YV,preds))
print("Root Mean Square Error:", rmse)
```
